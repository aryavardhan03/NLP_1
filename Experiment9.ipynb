```python
from gensim.models import word2vec
```

```python
import urllib.request
url="https://simple.wikipedia.org/wiki/Global_warming"
headerss={
    "user-agent":"mozilla/5.0"
}
req=urllib.request.Request(url,headers=headerss)
response=urllib.request.urlopen(req)
source=response.read()
```

```python
source
```

```python
soup=BeautifulSoup(source,'lxml')
```

```python
text=""
for paragraph in soup.find_all('p'):
  text+=paragraph.text + ' '
```

```python
text=re.sub(r'\[[0-9]*\]','',text)
text=re.sub(r'\s+',' ',text)
text=text.lower()
text=re.sub(r'\d','',text)
```

```python
sentences=nltk.sent_tokenize(text)
```

```python
sentences = nltk.sent_tokenize(text)

word_list = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sentences
```

```python
model=word2vec.Word2Vec(word_list,vector_size=100,min_count=1)
words=model.wv.index_to_key
```

```python
print(words)
```

```python
vector=model.wv['global']
```

```python
similar=model.wv.most_similar('global')
```
