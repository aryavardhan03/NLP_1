```python
import nltk
import re
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist

nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
text = """
Natural Language Processing is a field of Artificial Intelligence.
It focuses on enabling computers to understand human language.
Bag of Words is a simple and widely used text representation technique.
"""
```

```python
sentences = sent_tokenize(text)
print("Sentence Tokenization:")
for s in sentences:
    print(s)
```

```python
text_lower = text.lower()
print("\nLowercase Text:")
print(text_lower)
```

```python
clean_text = re.sub(r'\W', ' ', text_lower)

clean_text = re.sub(r'\s+', ' ', clean_text)

print("\nCleaned Text:")
print(clean_text)
```

```python
words = word_tokenize(clean_text)
print("\nWord Tokens:")
print(words)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in words if word not in stop_words]

print("\nFiltered Words (After Stopword Removal):")
print(filtered_words)
```

```python
freq_dist = FreqDist(filtered_words)

print("\nWord Frequency:")
for word, freq in freq_dist.items():
    print(word, ":", freq)
```

```python
plt.figure(figsize=(10,5))
freq_dist.plot(20)
plt.show()
```

```python
top_50_words = freq_dist.most_common(50)

print("\nTop 50 Frequent Words:")
for word, freq in top_50_words:
    print(word, ":", freq)
```
